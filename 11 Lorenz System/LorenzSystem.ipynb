```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import integrate
```

# 1.Lorenz System

We solve the three coupled differential equations

$$
\dot{x} = a(y-x)
$$
$$
\dot{y} = x(b-z)-y
$$
$$
\dot{z} = xy-cz\\
$$

Here $x,y,z $ are the coordinates and $a,b,c $ are coefficients

```python
def f(t,intialValues,a,b,c):
    x,y,z=intialValues
    dx=a*(y-x)
    dy=x*(b-z)-y
    dz=x*y-c*z
    return [dx,dy,dz]
```

### Chaotic solution: For b > 24.74
Two repulsive points exist

```python
a=10
b=50
c=8/3
t0=0
tFinal=50
nMax=5001 #number of points

#initial values.
x0=1
y0=1
z0=0
time=np.linspace(t0,tFinal,nMax)
chaotic=integrate.solve_ivp(f,[t0,tFinal],[x0,y0,z0],method='RK45',t_eval=time,args=(a,b,c))
```

```python
plotTraj=plt.axes(projection='3d')
plotTraj.plot3D(chaotic.y[0],chaotic.y[1],chaotic.y[2],linewidth=0.8)
#as we see, 2 repulsive points exist.
```

* Checking if it's a chaotic system

```python
chaotic2=integrate.solve_ivp(f,[t0,tFinal],[1.1*x0,y0,z0],method='RK45',t_eval=time,args=(a,b,c))
#making a little change in the initial values.
plotTraj=plt.axes(projection='3d')
plotTraj.plot3D(chaotic.y[0],chaotic.y[1],chaotic.y[2],linewidth=0.8)
plotTraj.plot3D(chaotic2.y[0],chaotic2.y[1],chaotic2.y[2],linewidth=0.8)
#although they starting position is almost identical (one above the other), however
#their ending position is very different (and we will compare  the values in the next cell)
```

```python
print(chaotic.y[:,-1]) 
print(chaotic2.y[:,-1])
#comparing the last 2 elements together.
```

### Non-chaotic solution: For b < 24.74
The two points become attractive


```python
a=10
b=20
c=8/3
t0=0
tFinal=50
nMax=5001 #number of points

#initial values.
x0=1
y0=1
z0=0
time=np.linspace(t0,tFinal,nMax)
nonChaotic=integrate.solve_ivp(f,[t0,tFinal],[x0,y0,z0],method='RK45',t_eval=time,args=(a,b,c))
```

```python
plotTraj=plt.axes(projection='3d')
plotTraj.plot3D(nonChaotic.y[0],nonChaotic.y[1],nonChaotic.y[2],linewidth=0.8)
#The two points become attractive.
```

* Checking if it's a chaotic system

```python
nonChaotic2=integrate.solve_ivp(f,[t0,tFinal],[1.1*x0,y0,z0],method='RK45',t_eval=time,args=(a,b,c))
#making a little change in the initial values.
plotTraj=plt.axes(projection='3d')
plotTraj.plot3D(nonChaotic.y[0],nonChaotic.y[1],nonChaotic.y[2],linewidth=0.8)
plotTraj.plot3D(nonChaotic2.y[0],nonChaotic2.y[1],nonChaotic2.y[2],linewidth=0.8)
#as we see they are almost identical
```

```python
print(nonChaotic.y[:,-1]) 
print(nonChaotic2.y[:,-1])
#almost identical
```
